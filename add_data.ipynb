```python
import pandas as pd
import math
from math import sin, cos, sqrt, atan2
import numpy as np
```

```python
df_2020 = pd.read_csv("202009-citibike-tripdata.csv")
```

```python
df_2020.head()
```

```python
def distance(lat1, lon1, lat2, lon2):
    # r = the earth's radius (km)
    r = 6378.137
    latx1 = math.radians(lat1)
    latx2 = math.radians(lat2)
    lon1 = math.radians(lon1)
    lon2 = math.radians(lon2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat2
    
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a),sqrt(1-a))
    
    d = r * c
    
    return d * 1000


```

```python
distance(40.8,-73.9,40.4,-74.0)
```

```python
df_2020["distance"] = np.vectorize(distance)(df_2020["start station latitude"],df_2020["start station longitude"],df_2020["end station latitude"],df_2020["end station longitude"])
```

```python
df_2020.head()
```

```python
df_2020["avg speed(m/s)"] = df_2020["distance"] / df_2020["tripduration"]
```

```python
df_2020.head()
```

```python
df_2020.to_csv("202009_data_for_tableau.csv")
```
